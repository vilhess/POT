```python
import numpy as np 
import matplotlib.pyplot as plt

from pot import spot, dspot
```

## Stationnary samples:

```python
train = np.random.standard_normal(1000)
stream = np.random.standard_normal(10_000)
```

### SPOT:

```python
ano, th = spot(init_set=train, stream_set=stream, init_seuil=0.98, proba=1e-4, n_points=10)

plt.figure(figsize=(15, 7))
plt.plot(stream)
plt.plot(th)
plt.scatter(ano, stream[ano])
plt.title('SPOT')
```

### DSPOT:

```python
ano, th, normalized, values = dspot(init_set=train, stream_set=stream, ws=10, init_seuil=0.98, proba=1e-4, n_points=10)

plt.figure(figsize=(15, 7))
plt.plot(stream)
plt.scatter(ano, stream[ano])
plt.title('DSPOT')
```

## Non-Stationnary samples:

```python
stream = np.linspace(1, 100, 1000)
stream = np.sin(0.5*stream) + 0.2*np.random.randn(1000)

train = stream[:300]
stream = stream[300:]
```

### SPOT:

```python
ano, th = spot(init_set=train, stream_set=stream, init_seuil=0.98, proba=1e-4, n_points=10)

plt.figure(figsize=(15, 7))
plt.plot(stream)
plt.plot(th)
plt.scatter(ano, stream[ano])
plt.title('SPOT')
```

### DSPOT:

```python
ano, th, normalized, values = dspot(init_set=train, stream_set=stream, ws=10, init_seuil=0.98, proba=1e-4, n_points=10)

plt.figure(figsize=(15, 7))
plt.plot(stream)
plt.scatter(ano, stream[ano])
plt.title('DSPOT')
```
